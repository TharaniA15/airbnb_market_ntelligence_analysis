```python
import os
import pandas as pd
from datetime import datetime
```

```python
def ingest_city_data(data_path):

    datasets = {}

    for file in os.listdir(data_path):

        if file.endswith(".csv"):

            file_name = file.replace(".csv", "")

            file_path = os.path.join(data_path, file)

            datasets[file_name] = pd.read_csv(file_path)

            print(f"{file} loaded successfully")

    return datasets
```

```python
def ingest_city_data_safe(data_path):

    datasets = {}

    for file in os.listdir(data_path):

        try:

            if file.endswith(".csv"):

                name = file.replace(".csv","")

                path = os.path.join(data_path,file)

                datasets[name] = pd.read_csv(path)

                print(f"{file} loaded successfully")


        except Exception as error:

            print(f"Error loading {file}: {error}")


    return datasets
```

## Fact and Dimension Tables

### Fact Table: fact_listing

The fact_listing table stores measurable business metrics and references dimension tables through foreign keys.

Columns:
- listing_id (Primary Key)
- host_id (Foreign Key)
- room_type_id (Foreign Key)
- location_id (Foreign Key)
- price
- total_reviews
- occupancy_rate
- estimated_revenue
- review_frequency

### Dimension Table: dim_host

Stores host information.

Columns:
- host_id (Primary Key)
- host_name

### Dimension Table: dim_room

Stores room type information.

Columns:
- room_type_id (Primary Key)
- room_type

### Dimension Table: dim_location

Stores geographical information.

Columns:
- location_id (Primary Key)
- neighbourhood
- neighbourhood_group

## Primary and Foreign Key Relationships

### Primary Keys

| Table | Primary Key |
|---------|---------|
| dim_host | host_id |
| dim_room | room_type_id |
| dim_location | location_id |
| fact_listing | listing_id |

### Foreign Keys

| Fact Table Column | References |
|---------|---------|
| host_id | dim_host.host_id |
| room_type_id | dim_room.room_type_id |
| location_id | dim_location.location_id |

These relationships create a star schema optimized for analytical querying and reporting.

```python
import time

def load_csv_with_retry(file_path,
                        retries=3):

    for attempt in range(retries):

        try:

            df = pd.read_csv(file_path)

            logging.info(
                f"Loaded {file_path}"
            )

            return df

        except Exception as e:

            logging.error(
                f"Attempt {attempt+1} failed: {e}"
            )

            time.sleep(2)

    raise Exception(
        f"Failed to load {file_path}"
    )
```

```python
listings_test = load_csv_with_retry(
    "../data/listings.csv"
)

print(
    listings_test.shape
)
```

```python
def ingest_city_data(config):

    logging.info(
        f"Starting ingestion for {config['city_name']}"
    )

    listings = load_csv_with_retry(
        f"{config['data_path']}/listings.csv"
    )

    reviews = load_csv_with_retry(
        f"{config['data_path']}/reviews.csv"
    )

    neighbourhoods = load_csv_with_retry(
        f"{config['data_path']}/neighbourhoods.csv"
    )

    return {
        "listings": listings,
        "reviews": reviews,
        "neighbourhoods": neighbourhoods
    }
```

## 3.5.2 Incremental Processing

```python
processed_ids = set()

print(
    "Tracker initialized"
)
```

```python
def get_new_records(df,
                    id_column):

    global processed_ids

    new_records = df[
        ~df[id_column].isin(
            processed_ids
        )
    ]

    processed_ids.update(
        new_records[id_column]
    )

    return new_records
```

### Incremental Processing Strategy

The pipeline tracks previously processed listing IDs.

During subsequent executions, only records with unseen IDs are processed.

This avoids unnecessary reprocessing and improves scalability for future data ingestion runs.

```python
import matplotlib.pyplot as plt

listing_master['price'].dropna().hist(
    bins=50
)

plt.title('Overall Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
listing_master[
    listing_master['price'] <= 1000
]['price'].hist(
    bins=50
)

plt.title(
    'Price Distribution (Price <= 1000)'
)

plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

### Finding

The overall price distribution is highly right-skewed. Most listings are concentrated below SGD 1,000, while a small number of premium properties exceed SGD 10,000.

### Business Interpretation

The Airbnb market is dominated by budget and mid-range accommodations, while a small premium segment targets luxury travellers. These premium listings significantly influence average price calculations.

### Business Action

Median price should be preferred over mean price when analysing market pricing, as extreme outliers distort average values.

```python
listing_master.boxplot(
    column='price',
    by='room_type',
    figsize=(10,5)
)

plt.title('Price Distribution by Room Type')
plt.suptitle('')
plt.xlabel('Room Type')
plt.ylabel('Price')

plt.show()
```

### Observation

The initial boxplot is heavily influenced by extreme price outliers. As a result, the central distribution of room-type prices is difficult to interpret.

A secondary visualization was created using median prices to better compare room-type pricing patterns.

```python
room_price = (
    listing_master
    .groupby('room_type')['price']
    .median()
    .sort_values(ascending=False)
)

room_price.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Median Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Median Price')

plt.show()
```

```python
top_neighbourhoods = (
    listing_master
    .groupby('neighbourhood')['price']
    .median()
    .sort_values(ascending=False)
    .head(10)
)

top_neighbourhoods.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 Neighbourhoods by Median Price')
plt.xlabel('Neighbourhood')
plt.ylabel('Median Price')

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
calendar['date'] = pd.to_datetime(
    calendar['date']
)

calendar['month'] = (
    calendar['date']
    .dt.month
)

monthly_availability = (
    calendar
    .groupby('month')['available']
    .apply(lambda x: (x == 't').mean() * 100)
)

monthly_availability
```

The calendar dataset contains a significant number of missing values in the price and adjusted_price fields. Therefore, robust temporal price trend analysis could not be performed using calendar-level pricing data.

```python
reviews['date'] = pd.to_datetime(
    reviews['date']
)

reviews['year'] = reviews['date'].dt.year
```

```python
review_trend = (
    reviews
    .groupby('year')
    .size()
    .reset_index(name='review_count')
)

review_trend
```

```python
review_trend.plot(
    x='year',
    y='review_count',
    kind='line',
    marker='o',
    figsize=(10,5)
)

plt.title('Review Volume Trend by Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')

plt.grid(True)

plt.show()
```

### Methodology Note

The dataset does not contain actual booking transactions. Therefore, review volume was used as a proxy measure for booking activity and demand trends.

```python
listing_master['host_category'] = pd.cut(
    listing_master['review_years'],
    bins=[-1, 1, 3, 100],
    labels=[
        'New Host (<1 Year)',
        'Growing Host (1-3 Years)',
        'Established Host (>3 Years)'
    ]
)

listing_master[
    ['review_years', 'host_category']
].head()
```

```python
host_price_comparison = (
    listing_master
    .groupby('host_category')
    .agg(
        median_price=('price', 'median'),
        listing_count=('id', 'count')
    )
)

host_price_comparison
```
